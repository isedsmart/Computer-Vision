I realized that rotating an image of different dimensions sizes can be very difficult since padding would have to be involved. So due to time, I wanted to first see if I could figure out how to rotate an image 180 degrees since I wouldn't run into any dimension errors.

```python
def rotateImg180Degrees(img):
    imageWidth = len(img[0])
    imageHeight = len(img)
    newimg = []
#     print(imageWidth)
#     print(imageHeight)
#     print(img)
    for y in range(imageHeight-1, 0, -1):
        tempimg = []
        for x in range(imageWidth-1, 0, -1):
            tempimg.append(img[y][x])
            # the new image will have appended all of the pixels in reverse order
        newimg.append(tempimg)
    return newimg
    
rotatedimage = rotateImg180Degrees(test_image)
# rotateImg180Degrees(test_image)

plt.imshow(rotatedimage)
plt.show()
```

```python
def padImg(img):
    imageWidth = len(img[0])
    imageHeight = len(img)
    if imageWidth > imageHeight:
        paddedImg = cv2.copyMakeBorder(img, int((imageWidth-imageHeight)/2), int((imageWidth-imageHeight)/2), 0, 0, cv2.BORDER_CONSTANT) # top, bottom, left, right
    else:
        paddedImg = cv2.copyMakeBorder(img, 0, 0, int((imageHeight-imageWidth)/2), int((imageHeight-imageWidth)/2), cv2.BORDER_CONSTANT)
    return paddedImg

print('This is the padded image')
plt.imshow(padImg(test_image))
plt.show()
```

```python
def rotateImg90Degrees(img):
    imageWidth = len(img[0])
    imageHeight = len(img)
    paddedimg = padImg(img)
    newimg = []
    for x in range(imageWidth-1, 0, -1):
        tempimg = []
        for y in range(imageHeight):
            tempimg.insert(0, paddedimg[y][x])
        newimg.append(tempimg)
    return newimg

plt.imshow(rotateImg90Degrees(test_image))
plt.show()
```

```python
def rotateImg270Degrees(img):
    imageWidth = len(img[0])
    imageHeight = len(img)
    paddedimg = padImg(img)
    newimg = []
    for x in range(imageWidth-1, 0, -1):
        tempimg = []
        for y in range(imageHeight-1, 0, -1):
            tempimg.insert(0, paddedimg[y][x])
        newimg.append(tempimg)
    return newimg

plt.imshow(rotateImg270Degrees(test_image))
plt.show()
```

I noted that there were two possibilities. One image could have a larger height while another could have a larger width. I decided to separate these two situations with an if statement. By doing so, it allowed me to cover both cases where if the image's width was longer than its height then the equation would be the image's height minus the image's width over 2. The reason why it's over 2 is because then it'll split the amount of pixels needed in half so that half the pixels can be added to the top and bottom of the image. Simiarly, if the image's height is longer than the image's width, then the equation becomes the image's width minus the image's height over 2. Then the pixels can be added on the left and right of the image.
