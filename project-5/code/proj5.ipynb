# Write Up

The student harris function is just the harris corner detector and the non maxima suppression in one function. It returns the x and y coordinates of the interest points that were found.

The get features function finds areas around the interest points (patches) and returns them. The size of the patch depends on the feature width number.

The student feature matching function calculates the cartesian distances of every individual feature between two feature vectors. Then calculates the confidence value of the matches by finding the ratio between the best 2 matches in each feature vector and runs them through a threshold. Lastly, it returns the matches and their confidence values.
